```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
import seaborn as sns
sns.set_color_codes("deep")
```

```python
df = pd.read_csv(r"D:\UNIVERSITY\Year 4, Term 1\STAT288\project\real project\final\cleaned_df.csv")
```

```python
best_worst_genres = ['War', 'History', 'Western', 'Horror', 'Thriller', 'Action', 'Drama', 'Comedy']
```

```python
filtered_df = df[df['genre name'].isin(best_worst_genres)]
```

```python
genre_profit = filtered_df.groupby('genre name')['profit'].mean().sort_values(ascending=False)
```

```python
genre_profit_millions = genre_profit / 10**7
```

```python
genre_profit_df = pd.DataFrame({
    'genre': genre_profit_millions.index,
    'profit_in_tens_of_millions': genre_profit_millions.values
})
```

```python
plt.barh(genre_profit_df["genre"],genre_profit_df["profit_in_tens_of_millions"],color=['g','grey','grey','grey','grey','r','r','r'])
plt.gca().invert_yaxis()
plt.xlabel("Average profit in 10's of millions ($)", fontsize=12)
plt.ylabel("Genre", fontsize=12)
plt.title("Average profit by Genre in 10's of millions ($)", fontsize=15)
plt.tight_layout()
plt.savefig("genre_profit_plot.png", format='png', dpi=300)
plt.show()

```
