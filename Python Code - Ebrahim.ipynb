```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("C:/Users/Ebrah/OneDrive/Desktop/cleaned_df_continent.csv")
```

# box plot for average voting by Continent

```python
#box plot assigning Continent to hue
plt.figure(figsize=(12, 8))
sns.boxplot(x="Continent", y="average voting", hue="Continent", data=df, palette="Set2", legend=False)

# Set title and labels
plt.title("Box Plot: Average Voting by Continent")
plt.xlabel("Continent")
plt.ylabel("Average Voting (Rating)")
plt.xticks(rotation=45)

# Save the plot first, then show it
plt.savefig(
    "C:/Users/Ebrah/OneDrive/Desktop/plots/box plot.png",
    bbox_inches='tight',
    dpi=200,
    transparent = False,
    facecolor = "white"
)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df=pd.read_csv("C:/Users/Ebrah/OneDrive/Desktop/movie_cast.csv")
```

```python
# Step 1: Replace 'not specified' with a random gender ('Male' or 'Female')
df['gender'] = df['gender'].apply(
    lambda x: np.random.choice(['Male', 'Female']) if x.strip().lower() == 'not specified' else x.strip().title()
)

# Step 2: Normalize gender values to ensure consistency (e.g., 'Male' and 'Female' in title case)
df['gender'] = df['gender'].replace({'male': 'Male', 'female': 'Female'})

# Step 3: Verify the update by checking the distribution of gender
gender_counts = df['gender'].value_counts()
print(gender_counts)

# Optional: Save the updated DataFrame to a new CSV
df.to_csv('updated_file.csv', index=False)
```

# pi chart for Gender Distribution of Lead Roles

```python
# Filter the data for order == 0 (lead roles)
lead_roles = df[df['order'] == 0]

# Count the gender distribution for lead roles
gender_counts = lead_roles['gender'].value_counts()

# Plot the gender distribution as a pie chart
plt.figure(figsize=(7, 7))
gender_counts.plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Gender Distribution of Lead Roles')
plt.ylabel('')  # Hide the y-label

# Save the plot first, then show it
plt.savefig(
    "C:/Users/Ebrah/OneDrive/Desktop/plots/pi chart.png",
    bbox_inches='tight',
    dpi=200,
    transparent = False,
    facecolor = "white"
)

plt.show()
```

# Top 10 Actors/Actresses by Lead Role Count

```python
# Count the number of lead roles per actor/actress
lead_role_count = lead_roles['name'].value_counts()

# Plot the top movies by lead role count (top 10 actors/actresses)
plt.figure(figsize=(10, 6))
lead_role_count.head(10).plot(kind='barh', color='skyblue')
plt.title('Top 10 Actors/Actresses by Lead Role Count')
plt.xlabel('Number of Lead Roles')
plt.ylabel('Actor/Actress Name')
plt.gca().invert_yaxis()  # To show the top on the top
# Save the plot first, then show it
plt.savefig(
    "C:/Users/Ebrah/OneDrive/Desktop/plots/H bar chart.png",
    bbox_inches='tight',
    dpi=200,
    transparent = False,
    facecolor = "white"
)

plt.show()
```

# Top 20 Actors/Actresses by Lead Role Count

```python
import matplotlib.pyplot as plt

# Count the lead role actors
lead_role_count = lead_roles['name'].value_counts()

# Get the top 20 actors/actresses by lead role count
top_actors = lead_role_count.head(20)

# Filter by gender for male and female
male_actors = lead_roles[lead_roles['name'].isin(top_actors.index) & (lead_roles['gender'] == 'Male')]
female_actors = lead_roles[lead_roles['name'].isin(top_actors.index) & (lead_roles['gender'] == 'Female')]

# Count the number of lead roles for male and female actors
male_count = male_actors['name'].value_counts().reindex(top_actors.index, fill_value=0)
female_count = female_actors['name'].value_counts().reindex(top_actors.index, fill_value=0)

# Plotting stacked bar chart
plt.figure(figsize=(10, 6))
plt.barh(top_actors.index, male_count, color='skyblue', label='Male')
plt.barh(top_actors.index, female_count, left=male_count, color='lightcoral', label='Female')

plt.title('Top 20 Actors/Actresses by Lead Role Count (Male vs Female)')
plt.xlabel('Number of Lead Roles')
plt.ylabel('Actor/Actress Name')
plt.gca().invert_yaxis()  # To show the top on the top
plt.legend()

# Save the plot first, then show it
plt.savefig(
    "C:/Users/Ebrah/OneDrive/Desktop/plots/top_20_bar_chart.png",
    bbox_inches='tight',
    dpi=200,
    transparent=False,
    facecolor="white"
)

plt.show()
```

# crew bar chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df=pd.read_csv("C:/Users/Ebrah/OneDrive/Desktop/movie_crew.csv")

# Bar Chart for Department Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='department', hue='department', palette='Set2', legend=False)
plt.title('Department Distribution')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(
    "C:/Users/Ebrah/OneDrive/Desktop/plots/crew bar chart.png",
    bbox_inches='tight',
    dpi=200,
    transparent = False,
    facecolor = "white"
)

plt.show()
```

# crew pi chart

```python
gender_counts = df['gender'].value_counts()
plt.figure(figsize=(8, 8))
gender_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette('Set3', len(gender_counts)), startangle=90)
plt.title('Gender Distribution')
plt.ylabel('')
plt.tight_layout()
plt.savefig(
    "C:/Users/Ebrah/OneDrive/Desktop/plots/crew pi chart.png",
    bbox_inches='tight',
    dpi=200,
    transparent = False,
    facecolor = "white"
)

plt.show()
```

# crew department by gender chart

```python
# Gender Distribution by Department (Grouped Bar Chart)
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='department', hue='gender', palette='Set2')
plt.title('Gender Distribution by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(
    "C:/Users/Ebrah/OneDrive/Desktop/plots/crew by gender chart.png",
    bbox_inches='tight',
    dpi=200,
    transparent = False,
    facecolor = "white"
)

plt.show()
```

# Heatmap for Departmen by Gender

```python
# Heatmap for Department vs Job vs Gender
gender_job_dept = df.groupby(['department', 'gender']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(gender_job_dept, annot=True, cmap='Blues', fmt='d')
plt.title('Heatmap for Department vs Gender')
plt.tight_layout()
plt.savefig(
    "C:/Users/Ebrah/OneDrive/Desktop/plots/crew heat map by gender.png",
    bbox_inches='tight',
    dpi=200,
    transparent = False,
    facecolor = "white"
)

plt.show()
```
